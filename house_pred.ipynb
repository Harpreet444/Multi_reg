```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv("C:\\Users\\hsd28\\Downloads\\Housing.csv")
df = data.drop(['mainroad','guestroom',	'basement',	'hotwaterheating',	'airconditioning',	'parking',	'prefarea',	'furnishingstatus'],axis=1)
df
```

```python
# import matplotlib.pyplot as plt

# Ensure df is defined and contains the required columns
features = ['area', 'bedrooms', 'bathrooms', 'stories']
target = 'price'

for feature in features:
    plt.figure()
    plt.scatter(df[feature], df[target], color='g', marker='+')
    plt.xlabel(feature)
    plt.ylabel(target)
    plt.title(f'Scatter Plot of {feature} vs {target}')
    plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','bathrooms','stories']],df[['price']])
```

```python
reg.predict(df[['area','bedrooms','bathrooms','stories']])
```
